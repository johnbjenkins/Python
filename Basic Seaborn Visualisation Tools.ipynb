# Seaborn Exercises

```python
# This exercise demonstrates the use of the Seaborn visualization library.

# It uses the Titanic dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
# Load dataset directly from Seaborn library
```

```python
titanic.head()
```

# Exercises

```python
# Create a jointplot of age v fare

sns.jointplot(x='fare',y='age',data=titanic)
```

```python
# Create a distplot showing the number of fares for a distinct value

sns.set_context('notebook')
sns.distplot(titanic['fare'],kde=False,bins=30,color='red')
```

```python
# Create a boxplot showing age distribution for each fare class

sns.boxplot(x='class',y='age',data=titanic)
```

```python
# Create a swarm plot showing age distribution for each fare class.

sns.swarmplot(x='class',y='age',data=titanic)
```

```python
# Create a countplot for sex of passenger

sns.countplot(x='sex',data=titanic)
```

```python
# Create a heatmap to show correlation between data categories

titcorr=titanic.corr()
```

```python
sns.heatmap(titcorr)
```

```python
# Create two plots, split by sex, to show the count of ages of passengers

g = sns.FacetGrid(data=titanic,col='sex')
g.map(sns.distplot,'age',kde=False,bins=10)
```
